```python
import pandas as pd
```

```python
df1=pd.read_csv("datasets/dataset_fpga.csv",encoding="utf-8")
df2=pd.read_csv("datasets/dataset_asic.csv",encoding="utf-8")
df3=pd.read_csv("datasets/dataset_vhdl.csv",encoding="utf-8")
```

```python
#Combine all dataframes
df=pd.concat([df1,df2,df3],axis=0).drop_duplicates()
df.shape
df.head()
```

```python
df['Title']=df['Title'].str.lower()
df = df.drop(df[df['Title'].str.find('stage')!=-1].index)
df.head()
```

```python
count=df.groupby(['Company']).count()['Title'].sort_values(ascending=False)[:20]
count.plot.bar(x='Companies',y='jobs');
```

```python
salary=df.groupby(['Company']).agg([ 'min', 'max'])['Salary_per_year']
```
